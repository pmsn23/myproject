```r
library(arules)
library(arulesViz)
PGBanklite<-readRDS("PGbanklite.rds")
PGBank<-readRDS("PGbank.rds")
```

```r
rules <- apriori(PGBank,parameter = list(minlen=2, supp=0.005, conf=0.8), appearance = list(rhs=c("response=Yes","response=No"), default="lhs"), control = list(verbose=F))
rules.sorted <- sort(rules, by="lift")
plot(rules.sorted, method="grouped")
inspect(rules.sorted, n=10)
```

```r
rules <- apriori(PGBanklite,parameter = list(minlen=2, supp=0.005, conf=0.8), appearance = list(rhs=c("response=Yes","response=No"), default="lhs"), control = list(verbose=F))
rules.sorted <- sort(rules, by="lift")
plot(rules.sorted, method="grouped")
inspect(rules.sorted, n=10)
```
